```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('Countries-exercise.csv')
```

```python
df
```

```python
x = df.iloc[:, 1:3]
x
```

```python
import os
os.environ['OMP_NUM_THREADS'] = '1'
```

```python
kmeans = KMeans(8)
```

```python
cluster = kmeans.fit(x)
```

```python
identified_clusters = kmeans.fit_predict(x)
identified_clusters
```

```python
cluster
```

```python
data_withcluster = df.copy() 
data_withcluster['clusters'] = identified_clusters
data_withcluster
```

```python
plt.scatter(data_withcluster['Longitude'], data_withcluster['Latitude'], c= data_withcluster['clusters'],cmap ='rainbow')
plt.xlim(-180,180)
plt.ylim(-90,90)
plt.show
```
