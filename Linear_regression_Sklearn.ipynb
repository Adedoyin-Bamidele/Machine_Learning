```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 
import seaborn as sns 
sns.set()
```

```python
data = pd.read_csv('1.04.+Real-life+example.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data['Price'] = data['Price'].fillna(data['Price'].mean())
```

```python
data.isnull().sum()
```

```python
data['EngineV'] = data['EngineV'].fillna(data['EngineV'].mean())
```

```python
data.describe()
```

```python
data = data.drop(columns = 'Model')
```

```python

a = data['Price'].quantile(0.7)

data['Above_75th_Percentile'] = data['Price'] > a

b = data.sort_values(by='Above_75th_Percentile', ascending=True)
b
```

```python
data= data.drop(columns = 'Above_75th_Percentile')
```

```python
data.head()
```

```python
data.describe(include = 'all')
```

```python
a = data['Price'].quantile(0.99)
data1 = data[data['Price']< a]
```

```python
data1.describe()
```

```python
b = data1['Mileage'].quantile(0.99)
data2 = data1[data1['Mileage']<b]
```

```python

c = data2['EngineV'].quantile(0.99)
data3 = data2[data2['EngineV']<c]
```

```python
d = data3['Year'].quantile(0.1)
df = data3[data3['Year']>d]
```

```python
df.describe(include = 'all')
```

```python
df.hist(bins = 500,  color='blue', edgecolor='black' , figsize = (12,11))
plt.show()
```

```python
f, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True, figsize =(15,3)) #sharey -> share 'Price' as y
ax1.scatter(df['Year'],df['Price'])
ax1.set_title('Price and Year')
ax2.scatter(df['EngineV'],df['Price'])
ax2.set_title('Price and EngineV')
ax3.scatter(df['Mileage'],df['Price'])
ax3.set_title('Price and Mileage')
```

```python

log_price = np.log(df['Price'])

# Then we add it to our data frame
df['log_price'] = log_price
df
```

```python
f, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True, figsize =(15,3)) #sharey -> share 'Price' as y
ax1.scatter(df['Year'],df['log_price'])
ax1.set_title('Price and Year')
ax2.scatter(df['EngineV'],df['log_price'])
ax2.set_title('Price and EngineV')
ax3.scatter(df['Mileage'],df['log_price'])
ax3.set_title('Price and Mileage')
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

variables = df[['Mileage','Year','EngineV']]


vif = pd.DataFrame()

vif["VIF"] = [variance_inflation_factor(variables.values, i) for i in range(variables.shape[1])]

vif["Features"] = variables.columns
```

```python
vif 
```

```python
data_n = df.drop(['Year'],axis=1)
```

```python
dummies = pd.get_dummies(data_n, drop_first=True)
```

```python
dummies
```

```python
df_dummies = dummies.astype(int)
```

```python
df_dummies.head()
```

```python
df_dummies = dummies.drop('Price', axis=1)
```

```python
df_dummies
```

```python
target = df_dummies['log_price']
```

```python
inputs = df_dummies.drop('log_price', axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(inputs)
```

```python
from sklearn.model_selection import train_test_split 
```

```python
x_train, x_test, y_train, y_test = train_test_split(inputs, target, test_size= 0.2 , random_state = 365)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linear = LinearRegression()
linear.fit(x_train, y_train)
```

```python
linear. score(x_train ,y_train)
```

```python
linear.coef_
```

```python
y_hat = linear.predict(x_train)
```

```python

plt.scatter(y_train, y_hat)
# Let's also name the axes
plt.xlabel('Targets (y_train)',size=18)
plt.ylabel('Predictions (y_hat)',size=18)

plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python

sns.distplot(y_train - y_hat)

plt.title("Residuals PDF", size=18)
```

```python
reg_summary = pd.DataFrame(inputs.columns.values, columns=['Features'])
reg_summary['Weights'] = linear.coef_
reg_summary
```

```python
y_hat_test = linear.predict(x_test)
```

```python
plt.scatter(y_test, y_hat_test, alpha=0.2)
plt.xlabel('Targets (y_test)',size=18)
plt.ylabel('Predictions (y_hat_test)',size=18)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
df_pf = pd.DataFrame(np.exp(y_hat_test), columns=['Prediction'])
df_pf.head()
```

```python
df_pf['Target'] = np.exp(y_test)
df_pf
```

```python
y_test = y_test.reset_index(drop=True)


y_test.head()
```

```python
df_pf['Target'] = np.exp(y_test)
df_pf
```

```python

df_pf['Residual'] = df_pf['Target'] - df_pf['Prediction']

df_pf['Difference%'] = np.absolute(df_pf['Residual']/df_pf['Target']*100)
df_pf
```
